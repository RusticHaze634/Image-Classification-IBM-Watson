# B. Food Classifier



````python
import json
from watson_developer_cloud import VisualRecognitionV3

Visual_Recognition = VisualRecognitionV3('2018-03-19', iam_apikey = 'enter_copied _api')

# url of a peach image
url = "https://www.shape.com/thmb/9SOyMDIovYw-qMVGgSJeYVEblqk=/750x0/filters:no_upscale():max_bytes(150000):strip_icc():gifv():format(webp)/peaches-promo-a0d430283f7440a1955a71c014d2628e.jpg"

result = visual_recognition.classify(url=url, classifier_ids =['food']).get_result()

print(json.dump(result, indent = 2))
````
